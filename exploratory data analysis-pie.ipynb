```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("animals_info new.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['Genus','Species','Kingdom','Subphylum','Class','Order','Family']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts().head(5)
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"Most {dt}")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("animals_info new.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Diet Types
# Split the Attributes
df['Attributes'] = df['Attributes'].str.split(',')

# Find Specific attributes & Count
df['Carnivore'] = df['Attributes'].apply(lambda x: 'Carnivore' in x)
df['Omnivore'] = df['Attributes'].apply(lambda x: 'Omnivore' in x)
df['Herbivore'] = df['Attributes'].apply(lambda x: 'Herbivore' in x)
comparison = {
    'Carnivore': df['Carnivore'].sum(),
    'Omnivore': df['Omnivore'].sum(),
    'Herbivore': df['Herbivore'].sum()
}

# Dataset
comparison_df = pd.DataFrame(list(comparison.items()), columns=['Diet', 'Count'])

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(comparison_df['Count'], labels=comparison_df['Diet'], autopct='%1.1f%%', startangle=90)
plt.title("Comparison of Diet Types")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("animals_info new.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Comparison of Activity Patterns
# Split the Attributes
df['Attributes'] = df['Attributes'].str.split(',')

# Find Specific attributes & Count
df['Nocturnal'] = df['Attributes'].apply(lambda x: 'Nocturnal' in x)
df['Diurnal'] = df['Attributes'].apply(lambda x: 'Diurnal' in x)
comparison = {
    'Nocturnal': df['Nocturnal'].sum(),
    'Diurnal': df['Diurnal'].sum(),
}

# Dataset
comparison_df = pd.DataFrame(list(comparison.items()), columns=['Activity Patterns', 'Count'])

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(comparison_df['Count'], labels=comparison_df['Activity Patterns'], autopct='%1.1f%%', startangle=90)
plt.title("Comparison of Activity Patterns")
plt.show()
```
